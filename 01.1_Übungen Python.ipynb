Schreiben sie einen Generator (Funktion mit "yield"), der aufsteigend ungerade Zahlen ab 20 ausgibt. Rufen sie diesen Generator in einer Schleife auf und speichern sie die ersten 10 Werte in einer Liste. Geben sie davon die Werte 4-7 aus.

```python
from collections import Counter
```

```python
def generate_ung(n):
    i = 21
    zaehler = 0
    while zaehler < n:
        zaehler += 1
        yield i
        i += 2
```

```python
liste = []
anzahl_ungerader_zahlen = 10
for i in generate_ung(anzahl_ungerader_zahlen):
    print(f"i: {i}")
    liste.append(i)
```

```python
print(liste)
```

```python
print(liste[3:7])
```

```python

Variante 2:


def generate():
    n = 20
    while True:
        yield n
        n += 1

data = generate()
unevens = (x for x in data if x % 2 != 0)

liste = []
for x in range(10):
    result = next(unevens)
    liste.append(result)
print(liste[3:7])
```

```python
def ungerade_zahlen():
    num = 19
    while True:
        num += 2
        yield num

gen = ungerade_zahlen()
erste_zehn = [next(gen) for _ in range(10)]
print(erste_zehn[3:7])
```
