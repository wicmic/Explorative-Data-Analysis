```python
from matplotlib import pyplot as plt
```

```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story", "LOTR - Rückkehr des Königs"]
num_oscars = [5, 11, 3, 8, 10, 11]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars, color="gold",
        label= ["0 - Annie Hall", "1 - Ben-Hur", "2 - Casablanca", "3 - Gandhi", "4 - West Side Story", "5 - LOTR - Rückkehr des Königs"])

plt.title("My Favorite Movies")     # add a title
plt.ylabel("# of Academy Awards")   # label the y-axis



plt.legend(loc=2)
plt.show()
```

```python
num_bins = 10                              # Anzahl Balken
grades = [12, 22, 25, 28, 32, 44, 52, 53, 56, 59, 43, 83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

plt.axis([-5, 105, 0, 5])                  # x-axis from -5 to 105,
                                           # y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)])    # x-axis labels at 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")

plt.hist(grades, num_bins, rwidth=0.95)
plt.show()
```
