```python
from matplotlib import pyplot as plt
```

```python

# Datensatz laden
blumen = []
zaehler = 0
with open('data/iris.tab') as f:
    for line in f:
        zaehler += 1
        if zaehler > 3:
            zeile = line.strip().split("\t")
            blumen.append([float(zeile[0]), float(zeile[1]), float(zeile[2]), float(zeile[3]), zeile[4]])
print(blumen)
from matplotlib import pyplot as plt

s_length = []
s_witdh = []
p_length = []
p_width = []
iris = []

for sepal_length, sepal_width, petal_length, petal_width, b_iris in blumen:
    s_length.append(sepal_length)
    s_witdh.append(sepal_width)
    p_length.append(petal_length)
    p_width.append(petal_width)
    iris.append(b_iris)
```

```python
plt.scatter(s_length[0:50], s_witdh[0:50], color="red", label="setosa")
plt.scatter(s_length[50:100], s_witdh[50:100], color="blue", label="versicolor")
plt.scatter(s_length[100:150], s_witdh[100:150], color="yellow", label="virginica")

plt.title("Irisarten Vergleich")  # add a title
plt.ylabel("Sepalweite")
plt.xlabel("Sepallänge")
plt.axis([0, 10, 0, 10])

plt.legend(loc=1)
plt.show()

```
