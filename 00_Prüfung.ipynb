```python
#Frage 7
pizza = [item[2]for item in content]
datum = [item[1]for item in content]

# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(datum)), pizza)

plt.title("Auslieferung Pizza")     # add a title
plt.ylabel("je Tag")   # label the y-axis

# label x-axis with movie names at bar centers
plt.xticks(range(len(datum)), pizza)

plt.show()
```
